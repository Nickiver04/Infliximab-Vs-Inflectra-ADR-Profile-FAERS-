```python
import numpy as np

all_reactions = reac[~reac['pt'].isin(exclude_terms)]['pt'].value_counts()

def calculate_signals(drug_reac, all_reac, drug_ids, total_ids):
    
    results = []
    drug_reac_counts = drug_reac['pt'].value_counts()
    
    total_drug_cases = len(drug_ids)
    total_all_cases = len(total_ids)
    
    for reaction, a in drug_reac_counts.items():
        
        A = a
        B = all_reac.get(reaction, 0) - A
        C = total_drug_cases - A
        D = total_all_cases - total_drug_cases - B
        
        if B <= 0 or C <= 0 or D <= 0:
            continue
        
        PRR = (A / (A + C)) / (B / (B + D))
        ROR = (A * D) / (B * C)
        
        se = np.sqrt(1/A + 1/B + 1/C + 1/D)
        ROR_lower = np.exp(np.log(ROR) - 1.96 * se)
        ROR_upper = np.exp(np.log(ROR) + 1.96 * se)
        
        results.append({
            'reaction': reaction,
            'A': A,
            'PRR': round(PRR, 3),
            'ROR': round(ROR, 3),
            'ROR_lower_95CI': round(ROR_lower, 3),
            'ROR_upper_95CI': round(ROR_upper, 3)
        })
    
    return pd.DataFrame(results)

signals_orig = calculate_signals(reac_orig_clean, all_reactions, orig_ids, valid_ids)
signals_bio = calculate_signals(reac_bio_clean, all_reactions, bio_ids, valid_ids)

sig_orig = signals_orig[(signals_orig['A'] >= 3) & (signals_orig['ROR_lower_95CI'] > 1)].sort_values('ROR', ascending=False)
sig_bio = signals_bio[(signals_bio['A'] >= 3) & (signals_bio['ROR_lower_95CI'] > 1)].sort_values('ROR', ascending=False)

print(f"Significant signals - Originator: {len(sig_orig)}")
print(f"Significant signals - Biosimilar: {len(sig_bio)}")

print("\nTop 10 signals - Originator:")
print(sig_orig.head(10))

print("\nTop 10 signals - Biosimilar:")
print(sig_bio.head(10))
```
